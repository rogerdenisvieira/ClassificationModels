# Preparação dos Dados

Importando as blibliotecas

```python
import numpy as np
import pandas as pd
import graphviz
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
from sklearn import tree

%matplotlib inline
```

Importando o dataset de churn disponibilizado pela IBM

```python
df = pd.read_csv("telco_churn.csv")
```

Aqui estou executando a seleção e o tratamento das features para valores aceitáveis pelos modelos:

```python
lbl = preprocessing.LabelEncoder()

for column_name in df.columns:
    if df[column_name].dtype == object:
        df[column_name] = lbl.fit_transform(df[column_name])
    else:
        pass
    
df.head()
```

Realizando o split do dataset em treino e validação.

```python
x_train, x_test, y_train, y_test = train_test_split(df, df["Churn"], test_size = 0.4)


df.head()
```

# Implementação dos Modelos

## Modelo de Árvore de Decisão

Criando um classificador e rodando a árvores de decisão para todas as features do dataset tendo o Churn como target da classificação


```python
clf = tree.DecisionTreeClassifier(random_state=111, criterion='entropy',max_depth=3,
                                   min_samples_leaf=3)
clf.fit(x_train, y_train)

print(clf.score(x_test, y_test))

print(confusion_matrix(y_test, clf.predict(x_test)))
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=df.columns,
                         class_names=df["Churn"],
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)
graph

```
